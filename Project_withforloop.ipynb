```python
import pandas as pd
import numpy as np

file1 = pd.read_csv('/Users/avapapetti/Desktop/CNV_Files/SL88823_20180802.cnv.csv', sep = "\t")
file2 = pd.read_csv('/Users/avapapetti/Desktop/CNV_Files/SL88824_20180802.cnv.csv', sep = "\t")

file1['Sample#'] = 1
file2['Sample#'] = 2

file1 = file1[file1["Chrom"] != 'chrM']
file2 = file2[file2["Chrom"] != 'chrM']
```

```python
common_cnvs = pd.DataFrame(columns = file1.columns)
```

```python
chrom_num = file1["Chrom"].unique()
```

```python
for i in range(len(chrom_num)):
    file1_cnvs = file1[(file1['Chrom'] == chrom_num[i])]
    file2_cnvs = file2[(file2['Chrom'] == chrom_num[i])]

    for j in range(len(file1_cnvs)):
        if(abs(file1_cnvs.iloc[j,1] - file1_cnvs.iloc[j,2]) >= 1000):
            for k in range(len(file2_cnvs)):
                if(abs(file2_cnvs.iloc[k,1] - file2_cnvs.iloc[k,2]) >= 1000):
                    if(abs(file1_cnvs.iloc[j,1] - file2_cnvs.iloc[k,1]) <= 25 and abs(file1_cnvs.iloc[j,2] - file2_cnvs.iloc[k,2]) <= 25):
                        if(file1_cnvs.iloc[j,3] == file2_cnvs.iloc[k,3]):
                             if(file1.iloc[j,4] >= .90 and file2.iloc[k,4] >= .90):
                                common_cnvs = pd.concat([common_cnvs, file1_cnvs.iloc[[j]], file2_cnvs.iloc[[k]]]).reset_index(drop = True)                
```
